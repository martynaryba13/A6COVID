### 3. Create a correlation matrix heatmap

```python
# create a subplot with matplotlib

f,ax = plt.subplots(figsize=(20,20))

# create correlation heatmap in seaborn by applying a heatmap onto correlation matrix and the subplots defined above
# 'annot' arguments allows plot to place the correlation coefficients onto the map

corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax) 
```

- 0: no relationship
- 0.1–0.3: weak relationship
- 0.3–0.5: moderate relationship
- 0.5–1.0: strong relationship

There are a few values that stand out on the graph. 
- Percent of people vaccinated with at least one dose and death rate per 100k has a correlation coeffcient of -0.61, which means a strong negative correlation. The higher the percent of people vaccinated the lower the death rate. 
- Similarly percent of people fully vaccinated and death rate per 100k has a correlation coefficient of -0.71, a strong negative correlation.
- There is a moderate negative correlation between case rate per 100k and percent full vaccinated at -0.38 and doses given per 100k at -0.48. This means that the more doses of the vaccine given, the lower the case rate. 
- There dose not seem to be any correlation between vaccinations given and hospitalizations. The correlation coefficients are -0.0038 and 0.0018 for percent of patients given at least one dose and fully vaccinated respectively.

### 6. Create a categorical plot and interpret the results.

```python
# create a histogram to visualize the distribution

sns.histplot(df_covid_merged['percent_with_at_least_one_dose'], bins = 20, kde = True)
```

```python
# create new column with vaccination level

df_covid_merged.loc[df_covid_merged['percent_with_at_least_one_dose'] < 65, 'vax_level'] = 'Average vaccination level'
```

```python
# create another subset

df_sub3 = df_covid_merged[['covid_cases_per_100k', 'death_rate_per_100k', 'vax_level']]
```

```python
df_sub3.head(5)
```

```python
# data set too large to run categorical plot, create sample

np.random.seed(4)
dev = np.random.rand(len(df_sub3)) <= 0.9
```

```python
dev
```

```python
# store 90% of dataframe as big

big = df_sub3[dev]
```

```python
# 10% of dataframe as small

small = df_sub3[~dev]
```

```python
small.shape
```

```python
big.shape
```

```python
# creat a categroical chart with small sample

sns.set(style="ticks")
g = sns.catplot(x="covid_cases_per_100k", y="death_rate_per_100k", hue="vax_level", data=small)
```

```python
g.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'vaxlevelplot.png'))
```

This categorical chart shows the relationship between COVID-19 case rate and death rate. We can interpret the chart as places with average vaccination level (marked in orange) have a higher death rate than those with a higher vaccination level.
